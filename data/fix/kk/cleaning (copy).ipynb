```python
df['nama'] = df['nama'].str.upper()
df['hubungan_kk'] = df['hubungan_kk'].str.upper()
df['jenis_kelamin'] = df['jenis_kelamin'].str.upper()
df['agama'] = df['agama'].str.upper()
df['pendidikan'] = df['pendidikan'].str.upper()
df['pekerjaan'] = df['pekerjaan'].str.upper()
df['status_kawin'] = df['status_kawin'].str.upper()
```

```python
df.head()
```

```python
import re
```

```python
def fun(x):
    c = re.sub('[aA-zZ][^\s]', '', x)
    c = re.sub(' ',"",c)
    c = re.sub('-',"/",c)
    c = re.sub(',',"",c)
    return c
```

```python
df['tgl_lahir'] = df['tgl_lahir'].apply(fun)
```
